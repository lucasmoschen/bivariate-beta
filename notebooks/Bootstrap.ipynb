```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import arviz as az
plt.style.use('science')

import sys
import os
sys.path.append('../scripts/python')

from parameter_estimation import BivariateBeta
```

```python
n = 50
alpha = (2,3,7,1)
U = np.random.dirichlet(alpha, size=n)
X = U[:,0] + U[:,1]
Y = U[:,0] + U[:,2]

distribution = BivariateBeta()
```

```python
alpha_estimates = distribution.bootstrap_method(X, Y, B=500, method=distribution.method_moments_estimator_1)
df = pd.DataFrame(alpha_estimates.T, columns=('alpha1', 'alpha2', 'alpha3', 'alpha4'))
```

```python
sns.pairplot(df)
```
